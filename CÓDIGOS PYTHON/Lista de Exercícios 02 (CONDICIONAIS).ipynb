Gasolina:
até 20 litros, desconto de 4% por litro
acima de 20 litros, desconto de 6% por litro 
Escreva um algoritmo que leia o número de litros vendidos, o tipo de combustível (codificado da seguinte forma: A-álcool, G-gasolina), calcule e imprima o valor a ser pago pelo cliente sabendo-se que o preço do litro da gasolina é R$ 4,50 o preço do litro do álcool é R$ 3,40.


```python
%reset -f

litros = float (input('Digite quantos litros deseja abastecer ? '))
tipo_combust = input('Digite o tipo de combústivel que deseja abastecer ? ')
preco_gas = 4.50
preco_alcool = 3.40

if litros <= 20 and tipo_combust == 'A' or 'a' :
  alcool_pagar1 = (litros - (litros * 3/100)) * preco_alcool
  print('Você deve pagar : R$',alcool_pagar1)
elif litros > 20 and tipo_combust == 'A' or 'a' :
  alcool_pagar2 = (litros - 5/100) * preco_alcool
  print('Você deve pagar : R$',alcool_pagar2)
elif litros <= 20 and tipo_combust == 'G' or 'g' :  
  gas_pagar1 = (litros - 4/100) * preco_gas
  print('Você deve pagar : R$',gas_pagar1)
elif litros > 20 and tipo_combust == 'G' or 'g' :  
  gas_pagar2 = (litros * 6/100) * preco_gas
  print('Você deve pagar : R$',gas_pagar2)
else :
  print('Valor inválido.')  
```
