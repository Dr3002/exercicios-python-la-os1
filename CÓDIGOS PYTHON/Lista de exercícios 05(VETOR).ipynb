3) Leia um vetor de 10 posições. Contar e escrever quantos valores pares ele possui.


```python
%reset -f

lista = [0] * 10
pares = 0
for i in range(10):
  lista[i] = int (input('Digite um número : '))
  if lista[i] % 2 == 0:
    pares = pares + 1
print('Esse vetor possui :',pares,'pares')

```

4) Escreva um programa que leia 10 números inteiros e os armazene em um vetor. Imprima o vetor, o maior elemento e a posição que ele se encontra.


```python
%reset -f

lista = [0] * 10
maior = 0
for i in range(10):
  lista[i] = int (input('Digite um número : '))
  if lista[i] > lista[maior]:
    maior = i
print(lista)
print('O maior elemento é ',lista[maior],' e a posição que ele se encontra é a ',maior + 1,'º',sep='')
```

6) Ler um vetor de 10 elementos. Crie um segundo vetor, com todos os elementos na ordem inversa, ou seja, o último elemento passará a ser o primeiro, o penúltimo será o segundo e assim por diante. Imprima os dois vetores. 


```python
%reset -f
tam = 10
vetor = [0] * tam
invertido = [0] * tam 
for i in range(tam):
  print('Digite o',i+1,'º valor :')
  vetor[i] = int (input())
for i in range(tam):
  invertido[tam-1-i] = vetor[i]

print('Vetor original :',vetor)
print('Vetor invertido :',invertido)
```

8) Dadas duas listas de tamanho N, faça um programa que diga se as mesmas possuem conteúdo igual. 

```python
%reset -f

N = int (input('Tamanho da lista : '))
lista1 = [0] * N
lista2 = [0] * N

print('Digite os elementos da lista 1 :')
for i in range(N):
  print('Digite o ',i+1,'º valor :  ',sep='',end='')
  lista1[i] = int (input())

print('Digite os elementos da lista 2 :')
for i in range(N):
  print('Digite o ',i+1,'º valor :  ',sep='',end='')
  lista2[i] = int (input())

iguais = True
for i in range(N):
  if lista1[i] != lista2[i]:
    iguais = False

print('Vetor 1:',lista1)
print('Vetor 2:',lista2)

if iguais:
  print('As listas são iguais.')
else:
  print('As listas são diferentes.')

  




```

9) Faça um programa que leia uma quantidade de números determinada pelo usuário e armazene em uma lista. Crie outras duas listas, uma para os valores pares e outra para os ímpares.
