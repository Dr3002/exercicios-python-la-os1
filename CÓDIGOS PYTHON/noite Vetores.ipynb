Solicite a idade de um grupo de 10 pessoas. Calcule a média de idade do grupo e informe quantas pessoas têm idade acima e abaixo da média.

```python
%reset -f

quant = 5
idades = [0] * quant
soma = 0
for i in range(quant):
  idades[i] = int(input('Informe a idade: '))
  soma = soma + idades[i]
media = soma / quant
contAcima = 0
contAbaixo = 0
for i in range(quant):
  if idades[i] > media:
    contAcima = contAcima + 1
  elif idades[i] < media:
    contAbaixo = contAbaixo + 1
print('Média de idade:',media)
print('Há',contAcima,'pessoas com idade acima da média e',contAbaixo,'pessoas com idade abaixo da média.')
```

Modifique a questão anterior para mostrar as idades acima e abaixo da média.

```python
%reset -f

quant = 5
idades = [0] * quant
soma = 0
for i in range(quant):
  idades[i] = int(input('Informe a idade: '))
soma = sum(idades)
media = soma / quant
print('Média de idade:',media)

print('Acima da média: ',end='')
for i in range(quant):
  if idades[i] > media:
    print(idades[i],end=' ')
print()
print('Abaixo da média: ',end='')
for i in range(quant):
  if idades[i] < media:
    print(idades[i],end=' ')

```

Faça um programa que leia números inteiros para uma lista. O usuário irá informar esses números até que informe zero (que não deve ser inserido na lista).
Imprima a lista e exclua todas as ocorrências de um valor informado pelo usuário.


```python
%reset -f

lista = []
numero = int(input('Digite um número (0 para encerrar): '))
while numero != 0:
  lista.append(numero)
  numero = int(input('Digite outro número (0 para encerrar): '))
print('Números informados:',lista)

numero = int(input('Que número deseja excluir? '))
#while numero in lista:
#  lista.remove(numero)

for i in range(lista.count(numero)):
  lista.remove(numero)

print('Lista depois da remoção:',lista)

```

Ler um conjunto de números reais, armazenando-o em vetor e calcular o quadrado das ´ componentes deste vetor, armazenando o resultado em outro vetor. Os conjuntos têm 10 elementos cada. Imprimir todos os conjuntos.


```python
%reset -f

lista = [0] * 10
quadrados = [0] * 10
for i in range(10):
  lista[i] = int(input('Informe o valor: '))
  quadrados[i] = lista[i] * lista[i] 
print('Números digitados:',lista)
print('Quadrados correspondentes:',quadrados)
```

Questão 4 lista 5

Escreva um programa que leia 10 números inteiros e os armazene em um vetor. Imprima o vetor, o maior elemento e a posição que ele se encontra.

```python
%reset -f
quant = 4
vetor = [0] * quant
maior_pos = 0
for i in range(quant):
  print('Digite o ',i+1,'º valor: ',sep='',end='')
  vetor[i] = int(input())
  if vetor[i] > vetor[maior_pos]:
    maior_pos = i
print('Valores digitados:',vetor)
print('O maior elemento é ',vetor[maior_pos],', e se encontra na posição ',maior_pos + 1,sep='')

```

Questão 8 lista 5:

Dadas duas listas de tamanho N, faça um programa que diga se as mesmas possuem conteúdo igual. 


```python
%reset -f

N = int(input('Qual o tamanho das listas? '))
lista1 = [0] * N
lista2 = [0] * N

print('Digite os elementos da lista 1:')
for i in range(N):
  print('Digite o ',i+1,'º valor: ',sep='',end='')
  lista1[i] = int(input())

print('Digite os elementos da lista 2:')
for i in range(N):
  print('Digite o ',i+1,'º valor: ',sep='',end='')
  lista2[i] = int(input())

if lista1 == lista2:
  print('Listas iguais.')
else:
  print('Listas diferentes.')

# refazer a questão usando laço para comparar posição a posição
```
