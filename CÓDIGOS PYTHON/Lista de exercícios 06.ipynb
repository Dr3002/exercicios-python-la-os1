5) Leia uma matriz de ordem n, informada pelo usuário. Calcule a soma dos elementos que estão acima da diagonal principal. 

```python
%reset -f

#INCOMPLETO

N = int (input('Digite a ordem da sua matriz :'))
matriz = []
for i in range(N):
  matriz.append([0]*N)

for i in range(N):
  for j in range(N):
    print('MATRIZ [',i+1,',',j+1,']:',sep='',end='')
    matriz[i][j] = int (input())

print('MATRIZ:')
for i in range(N):
  for j in range(N):
    print(matriz[i][j],end=' ')
  print()

soma = 0
for i in range(N):
  for j in range(N):
    soma = soma + matriz[i][N-2-j]

print('A soma dos elementos acima da diagonal principal é igual a ',soma)



# m[0,0] m[0,1] m[0,2] 
# m[1,0] m[1,1] m[1,2]
# m[2,0] m[2,1] m[2,2]

# n = 3
# j = 0 => n-2-j = 1
# j = 1 => n-2-j = 0
# j = 2 => n-2-j = -1
# j = 3 => n-2-j = -2
```

9) Leia uma matriz 5 x 10 que se refere respostas de 10 questões de múltipla escolha, referentes a 5 alunos. Leia também um vetor de 10 posições contendo o gabarito de respostas que podem ser a, b, c ou d. Seu programa deverá comparar as respostas de cada candidato com o gabarito e emitir um vetor denominado resultado, contendo a pontuação correspondente a cada aluno.


```python
%reset -f
alunos = 2
questoes = 3
respostas = []

for i in range(alunos):
  respostas.append([0]*questoes)

for i in range(alunos):
  print('Aluno ',i+1)
  for j in range(questoes):
    print('Questão',j+1,':',end=' ')
    respostas[i][j] = input()

gabarito = [0] * questoes
print('Gabarito :')
for i in range(questoes):
   print('Questão',i+1,':',end='')
   gabarito[i] = input()

resultado = [0] * alunos
for i in range(alunos):
  for j in range(questoes):
    if respostas[i][j] == gabarito[j]:
      resultado[i] = resultado[i] + 1  

for i in range(alunos):
  print('O aluno',i+1,'acertou',resultado[i],'questões')

```
