2) Faça um programa que leia um vetor de caracteres de tamanho N, contendo as respostas de um candidato em uma prova de concurso. A seguir, leia outro vetor, dessa vez contendo o gabarito da prova. Gere um terceiro vetor, de booleanos, indicando os erros e acertos do candidato. Com base nesse vetor, imprima o percentual de questões acertadas pelo candidato.

```python
%reset -f

def ler_respostas(tam,vet1):
  for i in range(tam):
    print('Questão ',i+1,'º : ',sep='',end='')
    vet1[i] = input()

def comp_respostas(tam,vet2,vet3,vetr):
  for i in range(tam):
    if vet2[i] == vet3[i]:
      vetr[i] = True
    else:
      vetr[i] = False

def perc_acert(tam,vetr):
  cont = 0
  for i in range(tam):
    if vetr[i] == True:
      cont = cont + 1
  return (cont/tam)*100
      
N = int (input('Digite o tamanho do vetor ? '))
respostas = [0] * N
gabarito = [0] * N
acertos = [0] * N

print()
print('ALUNO : ')
ler_respostas(N,respostas)
print()
print('GABARITO : ')
ler_respostas(N,gabarito)

comp_respostas(N,respostas,gabarito,acertos)
print()
print('O aluno acertou',perc_acert(N,acertos),'% da prova.')











  
```

5) Considere duas matrizes quadradas de ordem N. solicite uma linha e uma coluna e troque os valores da posição correspondente entre as matrizes. Imprima as duas matrizes antes e depois da troca.

```python
%reset -f

def criar_matriz(linhas,colunas):
  mat = []
  for i in range(linhas):
    mat.append([0]*colunas)
  return mat

def ler_matriz(mat,linhas,colunas):
  for i in range(linhas):
    for j in range(colunas):
      print('M[',i+1,',',j+1,'] :',sep='',end='')
      mat[i][j] = input()

def imprimir_matriz(mat,linhas,colunas):
   for i in range(linhas):
     for j in range(colunas):
       print(mat[i][j],end=' ')
     print() 

N = int (input('Qual a ordem da sua matriz ? '))
mat1 = criar_matriz(N,N)
mat2 = criar_matriz(N,N)

print('Matriz 1 :')
ler_matriz(mat1,N,N)
print('Matriz 2 :')
ler_matriz(mat2,N,N)

print('Matriz 1 :')
imprimir_matriz(mat1,N,N)
print('Matriz 2 :')
imprimir_matriz(mat2,N,N)

l = int (input('Qual linha deseja substituir o elemento ? '))
c = int (input('Qual coluna deseja substituir o elemento ?'))
l = l-1
c = c-1

aux = mat2[l][c]
mat2[l][c] = mat1[l][c]
mat1[l][c] = aux

print('Matriz 1 modificada:')
imprimir_matriz(mat1,N,N)
print('Matriz 2 modificada:')
imprimir_matriz(mat2,N,N)
```
