Faça um programa que solicite ao usuário a digitação de uma sequência de valores positivos. Para finalizar a sequência, o usuário digita um valor negativo. Calcule a média dos valores informados.

```python
%reset -f

quant = 0
soma = 0
valor = int(input('Digite um valor positivo: '))
while valor >= 0:
  soma = soma + valor
  quant = quant + 1
  valor = int(input('Digite um valor positivo: '))
media = soma / quant
print('A média dos valores digitados foi',media)
```

Questão 15 lista 3

Faça um programa que peça um número inteiro e determine se ele é ou não um número primo. Um número primo é aquele que é divisível somente por ele mesmo e por 1.


```python
%reset -f

num = int(input('Que número deseja saber se é primo? '))

primo = True
for i in range(2,num):
  if num % i == 0:
    primo = False

if primo:
  print('É primo.')
else:
  print('Não é primo.')
```

Questão 2 lista 4

Faça um programa para tabular a função: $f(x, y) = (x4 +3xy+y3 )/(2xy+3x+4y+2)$ para x = 2, 4, 6, 8 e y = 1, 3, 5, 7, 9, para cada valor de x. Devem ser impressos os valores de x, de y e de f(x,y).

```python
%reset -f

for x in range(2,9,2):
  for y in range(1,10,2):
    f = (x**4 + 3*x*y + y**3) / (2*x*y + 3*x + 4*y + 2)
    print('f(',x,',',y,') = ',f,sep='')
```

Faça um programa que imprima os N primeiros números primos. O usuário informa o valor de N.

```python
%reset -f

N = int(input('Quantos primos deseja imprimir? '))
# N = 8
# 1,2,3,5,7,11,13,17

numero = 1
cont_primos = 0
while cont_primos < N:
  primo = True
  for i in range(2,numero,1):
    if numero % i == 0:
      primo = False
  if primo:
    print(numero,end=' ')
    cont_primos = cont_primos + 1
  numero = numero + 1
```

Crie um algoritmo que simule o funcionamento de um caixa de supermercado. O caixa fica aberto até o fim do expediente e pode processar a compra de vários clientes. 
Cada cliente pode comprar vários itens. Ao ler cada item deve ser exibida uma mensagem para o operador do caixa perguntando se há mais itens a serem processados. Ao final, exiba quanto a compra custou ao cliente. E então solicite do operador do caixa a informação se deseja fechar o caixa. 
Quando o caixa for fechado, exiba quanto de dinheiro aquele caixa apurou no dia.


```python
%reset -f

fecharCaixa = 'n'
apurado = 0

while fecharCaixa == 'n': # vários clientes
  temItem = input('Tem itens para registrar? ')
  valorCompra = 0  
  while temItem == 's': #vários itens
    nomeItem = input('Nome: ')
    valorItem = float(input('Valor: R$ '))
    quantItem = float(input('Quantidade: '))
    valorCompra = valorCompra + valorItem * quantItem
    temItem = input('Há mais itens para registrar? ')
  print('Sua compra deu R$ ',valorCompra)
  apurado = apurado + valorCompra
  fecharCaixa = input('Deseja fechar o caixa? ')
print('Valor apurado no dia: R$',apurado)
```
