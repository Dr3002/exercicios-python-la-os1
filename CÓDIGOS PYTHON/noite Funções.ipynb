```python
%reset -f

def ler_respostas(vet,tam):
  for i in range(tam):
    print('Resposta questão',i+1,': ')
    vet[i] = input()

def comparar_respostas(vetr,vet1,vet2,tam):
  for i in range(tam):
    vetr[i] = vet1[i] == vet2[i]:

def calcular_percentual(vet,tam):
  cont = 0
  for i in range(tam):
    if vet[i]:
      cont = cont + 1
  return cont/tam*100

N = int(input('Quantas questões há na prova? '))
respostas = [0] * N
print('Respostas do candidato:')
ler_respostas(respostas,N)

gabarito = [0] * N
print('Gabarito:')
ler_respostas(gabarito,N)

acertos = [0] * N
comparar_respostas(acertos,respostas,gabarito,N)
print('O candidato acertou',calcular_percentual(acertos,N),'% da prova.')
```

```python
%reset -f

def criar_matriz(linhas,colunas):
  mat = []
  for i in range(linhas):
    mat.append([0] * colunas)
  return mat

def ler_matriz(mat,linhas,colunas):
  for i in range(linhas):
    for j in range(colunas):
      print('M[',i+1,',',j+1,']: ',sep='',end='')
      mat[i][j] = int(input())

def imprimir_matriz(mat,linhas,colunas):
  for i in range(linhas):
    for j in range(colunas):
      print(mat[i][j],end=' ')
    print()

N = int(input('Qual a ordem da matriz?'))
mat1 = criar_matriz(N,N)
mat2 = criar_matriz(N,N)
print('Matriz 1:')
ler_matriz(mat1,N,N)
print('Matriz 2:')
ler_matriz(mat2,N,N)
print()
print('Matriz 1:')
imprimir_matriz(mat1,N,N)
print()
print('Matriz 2:')
imprimir_matriz(mat2,N,N)

l = int(input('Quer trocar o elemento de que linha? '))
c = int(input('E que coluna? '))
l = l-1
c = c-1

aux = mat2[l][c]
mat2[l][c] = mat1[l][c]
mat1[l][c] = aux

print()
print('Depois da troca:')
print('Matriz 1:')
imprimir_matriz(mat1,N,N)
print()
print('Matriz 2:')
imprimir_matriz(mat2,N,N)
```

Faça um programa que leia uma matriz de dimensões informadas pelo usuário. Dado um valor, verifique quantas vezes algum múltiplo desse valor aparece na matriz.


```python
%reset -f

def criar_matriz(lin,col):
  mat = []
  for i in range(lin):
    mat.append([0] * col)
  return mat

def ler_matriz(mat,lin,col):
  for i in range(lin):
    for j in range(col):
      print('M[',i+1,',',j+1,']: ',sep='',end='')
      mat[i][j] = int(input())

def imprimir_matriz(mat,lin,col):
  for i in range(lin):
    for j in range(col):
      print(mat[i][j],end=' ')
    print()

def verificar_multiplos(mat,lin,col,valor):
  cont = 0
  for i in range(lin):
    for j in range(col):
      if mat[i][j] % valor == 0:
        cont = cont + 1
  return cont

linhas = int(input('Linhas: '))
colunas = int(input('Colunas: '))
matriz = criar_matriz(linhas,colunas)
ler_matriz(matriz,linhas,colunas)
imprimir_matriz(matriz,linhas,colunas)

valor = int(input('Qual o valor para encontrar os múltiplos? '))
print('Há',verificar_multiplos(matriz,linhas,colunas,valor),'múltiplos de',valor,'na matriz.')
```
