1. Faça um programa que solicite ao usuário o valor do litro de combustível (ex. 4,75) e quanto em dinheiro ele deseja abastecer (ex. 50,00). Calcule quantos litros de combustível o usuário obterá com esses valores. 

```python
%reset -f

dinheiro = float (input('Quantos reais irá gastar ?'))
gasolina = float (input('Qual o preço do litro?'))
VT = dinheiro // gasolina
print ('Você pode obter,', VT,'litros!')
```

2. Faça um programa que calcule a média de consumo (em km/l) de combustível de um veículo. O usuário deve informar o KM inicial (ex. 12500 km), o KM final (ex. 12700 km) e quantos litros foram gastos no percurso. 



```python
%reset -f

km_inicial = float (input('Qual o kilômetro inicial ?'))
km_final = float (input('Qual o kilômetro final ?'))
distancia_total = km_final - km_inicial 
litros_total = float (input('Quantos litros foram gastos no percurso ?'))
med_consu_per_litro = distancia_total // litros_total  * 10
print ('A média de consumo (em km/l) de combustível do seu carro é : ', med_consu_per_litro, 'KM/L !')

```

3. Faça um programa que calcule o valor a ser pago por uma dívida em atraso. O usuário deve informar o valor original da dívida (ex. R$ 50,00), a quantidade de dias em atraso (ex. 35 dias) e o valor da multa por dia de atraso (ex. R$ 0,25). 


```python
%reset -f

Valor_inicial = float (input('Qual o valor inicial da dívida atrasada ?'))
Dias_atraso = float (input('A quantos dias está com sua dívida em atraso ?'))
Valor_Multa_per_dia = float (input('Qual o valor da taxa por dia de atraso ?'))
Valor_Total_Multa_ = Valor_inicial + (Dias_atraso * Valor_Multa_per_dia / Valor_inicial)*100
print(f'Você deve pagar,' ,Valor_Total_Multa_, 'reais')
```

4. Faça um programa que calcule a área total (m2) de uma casa com 4 cômodos. O usuário deve inserir a largura e comprimento de cada um dos cômodos, calcular a área individual de cada um e por fim exibir a área total da casa. 


```python
%reset -f

sala_largura = float (input('Qual a largura da sala da sua casa ?'))
sala_comprimento = float (input('Qual o comprimento da sala da sua casa?'))
cozinha_largura = float (input('Qual a largura da cozinha da sua casa ?'))
cozinha_comprimento = float (input('Qual o comprimento da cozinha da sua casa?'))
copa_largura = float (input('Qual a largura da copa da sua casa ?'))
copa_comprimento = float (input('Qual o comprimento da copa da sua casa?'))
quarto_largura = float (input('Qual a largura do quarto da sua casa ?'))
quarto_comprimento = float (input('Qual o comprimento do quarto da sua casa?'))
A_sala = sala_largura * sala_comprimento
A_cozinha = cozinha_largura * cozinha_comprimento
A_copa = copa_largura * copa_comprimento
A_quarto = quarto_largura * quarto_comprimento
A_total_casa = A_sala + A_cozinha + A_copa + A_quarto
print ('A área total da casa é,',A_total_casa, 'M²')


```

5. Faça um programa que calcule a conversão monetária de Real para Dólar. O usuário informa o valor da cotação do dólar (ex.: 3,78) e quanto em real deseja converter (ex. 150,00). O programa exibe quantos dólares valem os reais informados.

6. Faça um programa que calcula o novo valor do salário de um funcionário. O usuário informa o salário atual (ex. 1250,00) e o percentual do reajuste (ex. 13,5 %). 

```python
%reset -f

sal_atual = float (input('Qual o valor do salário atual ?'))
perc_reaj = float (input('Qual o percentual do reajuste ?'))
reajuste = sal_atual * (perc_reaj/100)
novo_sal = sal_atual + reajuste
print ('O valor do novo salário é,',novo_sal, 'reais')
```

7. Faça um programa que calcula o tempo (em anos) que uma pessoa irá demorar para poupar R$ 1.000.000,00 (Um Milhão de Reais). O usuário informa o salário mensal e o total de despesas mensais. 


```python
%reset -f

sal_mensal = float (input('Qual o seu salário mensal?'))
desp_mensais = float (input('Qual o valor total das suas despesas mensais ?'))
resto_sal = sal_mensal - desp_mensais
din_poup_ano = resto_sal * 12 
temp_poup = (10**7 / din_poup_ano) / 10 
print ('Serão necessários:,',temp_poup,'anos para poupar R$1,000,000')
```
