Questão 1 lista 6

Leia uma matriz 4 x 4, conte e escreva quantos valores maiores que 10 ela possui.

```python
%reset -f

tam = 3
mat = []
for i in range(tam):
  mat.append([0] * tam)

for i in range(tam):
  for j in range(tam):
    print('M[',i+1,',',j+1,']: ',sep='',end='')
    mat[i][j] = int(input())

cont = 0
for i in range(tam):
  for j in range(tam):
    if mat[i][j] > 10:
      print(mat[i][j],end=' ')
      cont = cont + 1
print()
print('Foram',cont,'elementos maiores que 10.')
```

Questão 2 lista 6

Leia uma matriz 5 x 5. Leia também um valor X. O programa deverá fazer uma busca desse valor na matriz e, ao final, escrever a localização (linha e coluna) ou uma mensagem de “não encontrado”. 


```python
%reset -f

tam = 2
matriz = []
for i in range(tam):
  matriz.append([0] * tam)

for i in range(tam):
  for j in range(tam):
    print('M[',i+1,',',j+1,']: ',sep='',end='')
    matriz[i][j] = int(input())

for i in range(tam):
  for j in range(tam):
    print(matriz[i][j],end=' ')
  print()

encontrado = False
X = int(input('Digite o valor para buscar: '))
for i in range(tam):
  for j in range(tam):
    if matriz[i][j] == X:
      print('O elemento',X,'foi encontrado na linha',i+1,'e na coluna',j+1)
      encontrado = True

if not encontrado:
  print('Não encontrado.')
```

Faça um programa que leia uma matriz quadrada de ordem n. O programa deve mostrar a soma dos elementos da diagonal principal e da diagonal secundária, separadamente.

```python
%reset -f

n = int(input('Informe a ordem da matriz: '))

matriz = []
for i in range(n):
  matriz.append([0] * n)

for i in range(n):
  for j in range(n):
    print('M[',i+1,',',j+1,']: ',sep='',end='')
    matriz[i][j] = int(input())

for i in range(n):
  for j in range(n):
    print(matriz[i][j],end=' ')
  print()

soma = 0
for i in range(n):
  soma = soma + matriz[i][i]
print('A soma dos elementos da diagonal principal é',soma)

# m[0,0] m[0,1] m[0,2] m[0,3]
# m[1,0] m[1,1] m[1,2] m[1,3]
# m[2,0] m[2,1] m[2,2] m[2,3]
# m[3,0] m[3,1] m[3,2] m[3,3]

soma = 0
for i in range(n): 
  soma = soma + matriz[n-1-i][i]
print('A soma dos elementos da diagonal secundária é',soma)

# n = 4
# i = 0 => n-1-i = 3
# i = 1 => n-1-i = 2
# i = 2 => n-1-i = 1
# i = 3 => n-1-i = 0
```

Questão 9 lista 5

Leia uma matriz 5 x 10 que se refere respostas de 10 questões de múltipla escolha, referentes a 5 alunos. Leia também um vetor de 10 posições contendo o gabarito de respostas que podem ser a, b, c ou d. Seu programa deverá comparar as respostas de cada candidato com o gabarito e emitir um vetor denominado resultado, contendo a pontuação correspondente a cada aluno.


```python
%reset -f

alunos = 2
questoes = 3

respostas = []
for i in range(alunos):
  respostas.append([0] * questoes)

for i in range(alunos):
  print('Aluno',i+1,':')
  for j in range(questoes):
    print('Questão',j+1,':',end='')
    respostas[i][j] = input()

print('Gabarito:')
gabarito = [0] * questoes
for i in range(questoes):
  print('Questão',i+1,':',end='')
  gabarito[i] = input()

resultado = [0] * alunos
for i in range(alunos):
  for j in range(questoes):
    if respostas[i][j] == gabarito[j]:
      resultado[i] = resultado[i] + 1  

for i in range(alunos):
  print('O aluno',i+1,'acertou',resultado[i],'questões.')

```
