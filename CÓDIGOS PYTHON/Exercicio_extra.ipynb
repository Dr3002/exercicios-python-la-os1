16) Os números primos possuem várias aplicações dentro da Computação, por exemplo na Criptografia. Um número primo é aquele que é divisível apenas por um e por ele mesmo. Faça um programa que peça um número inteiro e determine se ele é ou não um número primo.

```python
%reset -f

num = int (input('Digite um número.'))

primo = True
for i in range(2,num):
  if num % i == 0:
    primo = False

if primo == True:
  print('É primo.')
else:
  print('Não é primo.') 

```
